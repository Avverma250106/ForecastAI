```python
sales.info()
sales['sale_date'] = pd.to_datetime(sales['sale_date'])
sales = sales.sort_values(by=['product_id','sale_date'])
```

```python
sales['day_of_week'] = sales['sale_date'].dt.weekday
sales['week_of_year'] = sales['sale_date'].dt.isocalendar().week.astype(int)
sales['month'] = sales['sale_date'].dt.month
sales['is_weekend'] = sales['day_of_week'].isin([5,6]).astype(int)
```

```python
LAGS = [1,7,14,28]
for lag in LAGS:
    sales[f'lag_{lag}'] = (
        sales.groupby('product_id')['quantity_sold'].shift(lag))
```

```python
WINDOWS = [7, 14, 30]

for window in WINDOWS:
    sales[f"rolling_{window}_mean"] = (
        sales.groupby("product_id")["quantity_sold"]
        .shift(1)
        .rolling(window)
        .mean()
    )
```

```python
online = online[
    (online["Quantity"] > 0) &
    (online["UnitPrice"] > 0)
]

online["InvoiceDate"] = pd.to_datetime(online["InvoiceDate"])
```

```python
online_features = (
    online.groupby("StockCode")
    .agg(
        avg_qty_per_order=("Quantity", "mean"),
        total_units_sold=("Quantity", "sum"),
        qty_std=("Quantity", "std"),
        avg_unit_price=("UnitPrice", "mean"),
        price_std=("UnitPrice", "std"),
        unique_customers=("CustomerID", "nunique")
    )
    .reset_index()
    .rename(columns={"StockCode": "product_id"})
)
```

```python
global_online_features = {
    "global_avg_qty_per_order": online["Quantity"].mean(),
    "global_qty_std": online["Quantity"].std(),
    "global_avg_unit_price": online["UnitPrice"].mean(),
    "global_price_std": online["UnitPrice"].std(),
    "global_unique_customers": online["CustomerID"].nunique()
}
```

```python
for col, val in global_online_features.items():
    sales[col] = val
```

```python
calendar = calendar[[
    "date", "wm_yr_wk", "event_name_1", "event_type_1", "snap_CA", "snap_TX", "snap_WI"
]]

calendar["date"] = pd.to_datetime(calendar["date"])

calendar["is_event"] = calendar["event_name_1"].notna().astype(int)
calendar["is_snap"] = (
    calendar[["snap_CA", "snap_TX", "snap_WI"]].max(axis=1)
)
```

```python
sales = sales.merge(
    calendar[["date", "wm_yr_wk", "is_event", "is_snap"]],
    left_on="sale_date",
    right_on="date",
    how="left"
)

sales.drop(columns=["date"], inplace=True)
```

```python
sell_prices = sell_prices.rename(columns={"item_id": "product_id"})
```

```python
sales["global_avg_price"] = sell_prices["sell_price"].mean()
sales["global_price_std"] = sell_prices["sell_price"].std()
sales["global_price_volatility"] = (
    sell_prices["sell_price"].std() /
    sell_prices["sell_price"].mean()
)
```

```python
sales["target"] = (
    sales.groupby("product_id")["quantity_sold"].shift(-1)
)
```

```python
final_df = sales.dropna().reset_index(drop=True)
```
